```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import OneHotEncoder

%matplotlib inline
```

```python
df = pd.read_csv('cleaneddata.csv')
```

```python
df = df.drop(columns=['waterfront','view', 'Unnamed: 0'])
```

```python
#one hot encoding zipcodes
zipcode = df['zipcode'].values.reshape(-1,1)


```

```python
encoder = OneHotEncoder(categories='auto').fit(zipcode)
```

```python
ohe = pd.DataFrame(encoder.transform(zipcode).toarray(),
                  columns=encoder.get_feature_names(['zipcode']))
```

```python
df_ohe = pd.concat([df.drop(columns='zipcode'), ohe], axis=1)
```

```python
df_ohe.to_csv(path_or_buf='clean_ohe.csv', index=False)
```
